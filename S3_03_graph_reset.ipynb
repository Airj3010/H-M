```python
from neo4j import GraphDatabase
import pandas as pd
import time

NEO4J_URI      = 'bolt://localhost:7687'
NEO4J_USER     = 'neo4j'
NEO4J_PASSWORD = 'password123'

driver = GraphDatabase.driver(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))

with driver.session() as s:
    print(s.run("RETURN 'Neo4j connected ✅' AS msg").single()['msg'])

def cypher(q, params=None):
    with driver.session() as s:
        return pd.DataFrame([r.data() for r in s.run(q, params or {})])

def run(q):
    with driver.session() as s:
        s.run(q)
```

```python
# Step 1 — Delete relationships first (the slow part)
print('Step 1: Deleting relationships first...')
total_rels = 0
with driver.session() as s:
    while True:
        deleted = s.run("""
            MATCH ()-[r]->()
            WITH r LIMIT 50000
            DELETE r
            RETURN COUNT(r) AS deleted
        """).single()['deleted']
        total_rels += deleted
        if deleted > 0:
            print(f'  Relationships deleted: {total_rels:,}')
        else:
            break
print(f'  ✅  All relationships deleted: {total_rels:,}')

# Step 2 — Now delete nodes (fast because no relationships left)
print('\nStep 2: Deleting nodes...')
total_nodes = 0
with driver.session() as s:
    while True:
        deleted = s.run("""
            MATCH (n)
            WITH n LIMIT 50000
            DELETE n
            RETURN COUNT(n) AS deleted
        """).single()['deleted']
        total_nodes += deleted
        if deleted > 0:
            print(f'  Nodes deleted: {total_nodes:,}')
        else:
            break
print(f'  ✅  All nodes deleted: {total_nodes:,}')
print('\n  ✅  Graph completely empty — safe to reload')
```
